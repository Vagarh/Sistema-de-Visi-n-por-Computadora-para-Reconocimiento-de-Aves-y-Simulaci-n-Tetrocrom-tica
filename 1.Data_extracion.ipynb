```python
import os
import tarfile
from zipfile import ZipFile
import glob

# 1) Define rutas
BASE_DIR = "/content/drive/MyDrive/Proyecto_Curso_BEIT_aves"
DEST    = os.path.join(BASE_DIR, "birds_image")
ZIP_PATH = os.path.join(BASE_DIR, "archive.zip")

# 2) Crea carpeta destino
os.makedirs(DEST, exist_ok=True)

# 3) Descomprime el ZIP original
with ZipFile(ZIP_PATH, 'r') as zip_ref:
    zip_ref.extractall(DEST)

# 4) Lista los archivos resultantes
print("Archivos en birds_image tras descomprimir ZIP:")
for f in os.listdir(DEST):
    print(" -", f)

# 5) Extrae todos los .tgz en subcarpetas con el mismo nombre (sin extensión)
for tgz_path in glob.glob(os.path.join(DEST, "*.tgz")):
    nombre = os.path.splitext(os.path.basename(tgz_path))[0]
    subdir = os.path.join(DEST, nombre)
    os.makedirs(subdir, exist_ok=True)
    with tarfile.open(tgz_path, 'r:gz') as tar:
        tar.extractall(path=subdir)
    print(f"→ Extraído {os.path.basename(tgz_path)} en {nombre}/")

# 6) (Opcional) Verifica contenido de una de las subcarpetas
print("\nEjemplo de contenido en la subcarpeta 'CUB_200_2011':")
for f in os.listdir(os.path.join(DEST, "CUB_200_2011"))[:10]:
    print("  •", f)

```
